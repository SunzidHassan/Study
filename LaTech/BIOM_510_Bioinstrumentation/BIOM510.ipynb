```python
import numpy as np
import matplotlib.pyplot as plt

# 1) Define parameters
omega0 = 2e4  # Example: angular cutoff frequency (adjust for your case)
t_min = -0.006
t_max = 0.006
N_points = 1000

# 2) Create time array
t = np.linspace(t_min, t_max, N_points)

# 3) Define the function v(t).
#    This is just a sample sinc formâ€”replace with whatever expression you have!
#    Note: np.sinc(x) in Python is normalized so that sinc(x) = sin(pi*x)/(pi*x).
#    If your formula is different, adjust accordingly.
def v_of_t(t):
    # Example: v(t) = (omega0 / pi) * sinc( (omega0/pi) * t )
    return (omega0 / np.pi) * np.sinc((omega0/np.pi) * t)

v = v_of_t(t)

# 4) Plot
plt.figure(figsize=(6,4))
plt.plot(t, v, label='v(t)')
plt.xlabel('Time (s)')
plt.ylabel('v(t)')
# plt.title('Impulse Response')
plt.grid(True)
# plt.legend()
plt.show()
```
