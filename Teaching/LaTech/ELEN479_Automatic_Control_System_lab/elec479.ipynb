### Closed-loop or Feedback-control System
A control system with an open loop as its forward path and one or more feedback loops or paths between its output and input. Some portion of the output is returned 'back' to the input. This allows reduction of error (desired output - actual output) and managing disturbance.

Automatic control systems are closed loop systems, that tries to minimize error.

### Transfer function
Establishes relation of input and output of a Linear, time-invariant system.

Let's assume that is a car suspension system, input bumps are $u(t)$ and output car motion are $y(t)$. This can be modeled using a 4th order differential equation. Given a bump, the graph of the equation will tell us the max travel and duration of travel of the car - all of which depend on time.

Now, it is possible to convert the differential equation into a `Transfer function` using `Laplace transform`. This transforms the differential equation in the time domain into a transfer function in complex plane of $s=\sigma+j\omega$ domain, where the imaginary component $\omega$ is the frequency of the signal, and the real component $\sigma$ is the exponential growth/decay of the signal.

If we have the input as a function of $s$ or in terms of exponential growth/decay and frequency of a signal, then we can transfer it by multiplying it with transfer function $$u(s).\frac{y(s)}{u(s)}=y(s)$$. In this case, the transfer function is the ratio of the output and input in the $s$ domain.

If there are a series of differential equations in time-domain that represent the system, where output of one equation is the input of another equation, then solving the whole system requires computation heavy covolution calculations. But in $S$ domain they convert to multiplication operations. We can also understand about the stability and system response information from the locations of `gains`, `poles` and `zeros` of the `Transfer function`. `Transfer functions` also allow easy manipulation in `Block diagram` form.

In a `Transfer function`, `Poles` are values of $s$ that cause $G(s)\rightarrow\infty$, and `Zeros` are values of $s$ that cause $G(s)\rightarrow 0$. For example, for $$G(s)=\frac{s+5}{(s+1)(s+3)}$$, `Poles` are $s=-1$ and $s=-3$ and `Zeros` are $s=-5$, and `Gain` is the value of $G(s)$ or the input-output ratio under steady-state conditions, or the value of the function when $s=0$.

For example, if any `Pole` is the right-half of the complex domain, that is if $\sigma$ has any positive value, then the system is unstable. In a series, if any `Transfer function` is unstable, then the entire system is unstable. For example, this is a stable series $$\rightarrow\left[\frac{s}{s+0.3}\right]\rightarrow\left[\frac{1}{s+1}\right]\rightarrow$$. Additionally, in a feedback-system by solving the closed-loop transfer function and checking the locations of the `Poles`. We can also design `Gains`, `Poles` and `Zeros` to achieve desired characteristic in our system.


### Block diagrams
* Block Diagrams are pictorial representation of a `Control system`.
* Block diagram representation is used to build a mathematical representation of a control system which cam be emulated on a computer.
* Block diagram representation is used to calculate the overall transfer function of the system.

Elements of a Block diagram:
$$\overset{R(s)}{\rightarrow}\left[G(s)\right]\overset{C(s)}{\rightarrow}$$
* The signal into the block represents the input $R(s)$
* The signal out of the block represents the output $C(s)$.
* The block itself represents the transfer function $G(s)$.

#### Summing Point:
A summing point/junction in a block diagram represents the dynamic summation of two (or more) signals.

Take-off point in a block diagram represents a point where the signal branches out and goes concurrently to the other blocks or summing points.

#### Reduction of Multiple Subsystems
* In order to calculate the output w.r.t. any given input, we need the overall transfer function of the system.
* When the system is complex (having multiple subsystems), we need to reduce the complex block diagram representation into a single block.
* The reduction is done using Block diagram algebra.
